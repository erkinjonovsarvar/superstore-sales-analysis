Finally, I will display the head of the `Dim_Date` DataFrame to verify its structure.



```python
Dim_Date = df[['Order Date', 'Year', 'Month', 'Quarter', 'DayOfWeek', 'WeekOfYear']].drop_duplicates().sort_values('Order Date').reset_index(drop=True)
Dim_Date['Date Key'] = Dim_Date['Order Date'].dt.strftime('%Y%m%d').astype(int)

# Reorder columns to place 'Date Key' first
Dim_Date = Dim_Date[['Date Key', 'Order Date', 'Year', 'Month', 'Quarter', 'DayOfWeek', 'WeekOfYear']]

print("Dim_Date table created successfully with unique date information.")
print(Dim_Date.head())
```

```python
Fact_Sales = df[['Order ID', 'Customer ID', 'Product ID', 'Order Date', 'Ship Date', 'ShipDuration', 'Sales', 'Quantity', 'Discount', 'Profit', 'Profit Margin', 'DayOfWeek', 'WeekOfYear', 'Quarter', 'Year', 'Month']]

print("Fact_Sales table created successfully with transactional data.")
print(Fact_Sales.head())
```

**Reasoning**:
I need to construct the `Fact_Sales` table by selecting the specified relevant columns from the main DataFrame `df` to create a `fact_sales_base` DataFrame. Then, I will merge this base with `Dim_Date` to incorporate the `Date Key`, drop the original `Order Date` column, reorder all columns to place foreign keys first, and finally display the head of the resulting `Fact_Sales` DataFrame to verify its structure.



```python
fact_sales_base = df[['Order ID', 'Customer ID', 'Product ID', 'Postal Code', 'Ship Mode', 'Sales', 'Quantity', 'Discount', 'Profit', 'Profit Margin', 'ShipDuration', 'Order Date']].copy()

Fact_Sales = pd.merge(fact_sales_base, Dim_Date[['Order Date', 'Date Key']], on='Order Date', how='left')

Fact_Sales.drop(columns=['Order Date'], inplace=True)

# Reorder columns
Fact_Sales = Fact_Sales[[
    'Date Key', 'Customer ID', 'Product ID', 'Postal Code', 'Ship Mode', 'Order ID',
    'Sales', 'Quantity', 'Discount', 'Profit', 'Profit Margin', 'ShipDuration'
]]

print("Fact_Sales table created successfully with transactional data linked to Date Key.")
print(Fact_Sales.head())
```

```python
# To save a plot, use plt.savefig('filename.png', bbox_inches='tight') before plt.show()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

df['Order Date'] = pd.to_datetime(df['Order Date'])
```

1. PROFIT vs DISCOUNT


```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Discount', y='Profit', alpha=0.5)
sns.regplot(data=df, x='Discount', y='Profit', scatter=False, color='red')
plt.title("Profit vs Discount")
plt.show()

# correlation
print("Correlation:", df[['Discount','Profit']].corr())

# bin analysis
df['discount_bin'] = pd.cut(df['Discount'], bins=5)

discount_analysis = df.groupby('discount_bin').agg({
    'Sales':'sum',
    'Profit':'sum'
})

print(discount_analysis)
plt.savefig("profit_vs_discount.png", bbox_inches='tight')
```

2. CATEGORY ANALYSIS

```python
category = df.groupby('Category').agg({
    'Sales':'sum',
    'Profit':'sum'
}).sort_values(by='Profit', ascending=False)

print(category)

category.plot(kind='bar', figsize=(8,5))
plt.title("Category Performance")
plt.show()
save_plot("profit_vs_discount")
```


3. SUB-CATEGORY (TOP & WORST)





```python
subcat = df.groupby('Sub-Category').agg({
    'Sales':'sum',
    'Profit':'sum'
}).sort_values(by='Profit', ascending=False)

top_10 = subcat.head(10)
worst_10 = subcat.tail(10)

print("TOP 10:\n", top_10)
print("\nWORST 10:\n", worst_10)
```
